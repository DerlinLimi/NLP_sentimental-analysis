```python
def fn(col):
    corpus = []
    for i in col:
        cmt = re.sub('[^a-zA-Z]',' ',str(i))
        cmt = cmt.lower()
        cmt = cmt.split()
        cmt = [lm.lemmatize(word) for word in cmt if word not in set(stopwords.words('english'))]
        corpus.append(' '.join(str(x) for x in cmt))
    return corpus
```

```python
corpus = fn(data['Comment'])
```

```python
cv = CountVectorizer(ngram_range=(1,2))
traindata = cv.fit_transform(corpus)
x = traindata
y = data.Emotion
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
y_pred_mnb=mnb.predict(x_test)
y_pred_mnb
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
print("MultinomialNB accuracy score:\t",metrics.accuracy_score(y_test,y_pred_mnb))
print("MultinomialNB training_data accuracy score:\t",mnb.score(x_train,y_train))
print("MultinomialNB testing_data accuracy score:\t",mnb.score(x_test,y_test))
```

```python
cm_mnb=confusion_matrix(y_test,y_pred_mnb)
cm_mnb
```

```python
print("MultinomialNB Classification Report:\n\n",metrics.classification_report(y_test,y_pred_mnb))
```

```python
# ['sadness', 'anger', 'love', 'surprise', 'fear', 'joy'],
def expression_check(prediction_input):
    if prediction_input == 0:
        print("Input statement has ANGER Sentiment.")
    elif prediction_input == 1:
        print("Input statement has FEAR Sentiment.")
    elif prediction_input == 2:
        print("Input statement has JOY Sentiment.")
    elif prediction_input == 3:
        print("Input statement has LOVE Sentiment.")
    elif prediction_input == 4:
        print("Input statement has SADNESS Sentiment.")
    elif prediction_input == 5:
        print("Input statement has SURPRISE Sentiment.")
    else:
        print("Invalid Statement.")
```

```python
def sentiment_predictor(input):
    input = fn(input)
    transformed_input = cv.transform(input)
    prediction = lr.predict(transformed_input)
    expression_check(prediction)
```
